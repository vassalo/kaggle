```python
import tensorflow as tf
import pandas as pd
import matplotlib.image as mpimg
import os
import numpy as np
import seaborn as sn

from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense
```

```python
train_df = pd.read_csv('data/train.csv')

train_imgs = []
y_train = []
for r, d, f in os.walk('data/train'):
    for file in f:
        train_imgs.append(mpimg.imread(os.path.join(r, file)))
        y_train.append(int(train_df.loc[train_df['id'] == file]['has_cactus']))

test_ids = []
test_imgs = []
for r, d, f in os.walk('data/test'):
    for file in f:
        test_ids.append(file)
        test_imgs.append(mpimg.imread(os.path.join(r, file)))
        
x_train = np.array(train_imgs)
x_test = np.array(test_imgs)

x_train = x_train / 255.
x_test = x_test / 255.


# Creates the model
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5, 5), padding='Same', activation='relu', input_shape=(32, 32, 3)))
model.add(Conv2D(filters=32, kernel_size=(5, 5), padding='Same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Flatten(input_shape=(32, 32, 3)))
model.add(Dense(512*3, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(2, activation="softmax"))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train
model.fit(x_train, y_train, epochs=20)


# Predict and save
predictions = np.argmax(model.predict(x_test), axis=1)
predictions = pd.DataFrame(data=[[test_ids[i], predictions[i]] for i in range(len(predictions))],
                           columns=['id', 'has_cactus'])
print(predictions)
predictions.to_csv('data/guess.csv', index=False)
```
